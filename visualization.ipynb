```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('Parkinsson disease.csv')
```

```python
numeric_df = df.drop('name',axis=1)
```

```python
import math

num_plots = len(numeric_df.columns)
ncols = 3
nrows = math.ceil(num_plots / ncols)


fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows * 4))


for col, ax in zip(numeric_df.columns, axes.flatten()):
    sns.kdeplot(data=numeric_df, x=col, fill=True, ax=ax)
    ax.set_title(f'Distribution of {col}')


for i in range(num_plots, len(axes.flatten())):
    axes.flatten()[i].set_visible(False)


plt.tight_layout()
plt.show()

```

```python
sns.countplot(data=numeric_df , x = 'status',color = 'blue')
```

```python
corr_matrix = numeric_df.corr()
plt.figure(figsize=(18, 15))
sns.heatmap(corr_matrix, cmap='coolwarm',annot=True)
plt.title('Correlation Matrix of Parkinson\'s Disease Features')
plt.show()
```
